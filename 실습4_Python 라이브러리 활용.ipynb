### 3D surface 그리기
- ref: https://matplotlib.org/stable/gallery/mplot3d/surface3d.html

```python
# fig = plt.figure()
# ax = fig.add_subplot(projection='3d')

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}) # 위 코드와 같은 동작

#함수 선언
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))
    
# 데이터 생성
# x축, y축 생성 후 meshgrid를 사용해 XY 평면 좌표 생성
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

# 함수를 통해 XY평면 좌표의 각 값에 대해 Z값 계산
Z = f(X, Y)

# surface plot 그리기
surf = ax.plot_surface(X, Y, Z)
```

#### 참고: Colormap을 사용하여 값에 따라 색칠하기
- color preset: https://matplotlib.org/stable/tutorials/colors/colormaps.html

```python
from matplotlib import cm
from matplotlib.ticker import LinearLocator

fig, ax = plt.subplots(subplot_kw={"projection": "3d"}) 

# 데이터 생성
# x축, y축 생성 후 meshgrid를 사용해 XY 평면 좌표 생성
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

# 함수를 통해 XY평면 좌표의 각 값에 대해 Z값 계산
Z = f(X, Y)

# Plot the surface, cmap 설정
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```

### 3D Quiver 그리기 (벡터장)
- ref: https://matplotlib.org/stable/gallery/mplot3d/quiver3d.html

```python
ax = plt.figure().add_subplot(projection='3d')

# 각 벡터의 시작 지점 좌표 설정
x, y, z = np.meshgrid(np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.2),
                      np.arange(-0.8, 1, 0.8))

# 각 벡터의 방향 설정
u = np.sin(np.pi * x) * np.cos(np.pi * y) * np.cos(np.pi * z)
v = -np.cos(np.pi * x) * np.sin(np.pi * y) * np.cos(np.pi * z)
w = (np.sqrt(2.0 / 3.0) * np.cos(np.pi * x) * np.cos(np.pi * y) *
     np.sin(np.pi * z))

# 벡터장 그리기, length: 벡터의 길이, normalize: 벡터의 길이를 모두 같게 설정
ax.quiver(x, y, z, u, v, w, length=0.1, normalize=True)

plt.show()
```

### 수식을 변수에 저장하여 수식 사용하기

```python
expr = x ** 2 + 1
expr
```

- `subs()` 를 통해 기호 변수에 값을 대입하거나 다른 기호 변수로 대체 가능

```python
expr.subs(x, 2)
```

```python
expr.subs(x, z)
```

```python
expr = (x + y) ** 2 + z
expr
```

```python
expr.subs([(x, 1), (y, 2), (z, 3)])
```

### 도함수 구하기
- `sympy.diff()` 함수를 이용하여 도함수를 구할 수 있음
- ref: https://docs.sympy.org/latest/tutorial/calculus.html

```python
sp.diff(sp.cos(x), x)
```

```python
sp.diff(x**4, x, 3) # 3번 미분
```

```python
expr = sp.exp(x*y*z)
sp.diff(expr, x, y, z, z) # 편미분도 가능
```

### 참고: 적분
- `sympy.integrate()` 함수 이용

```python
sp.integrate(sp.cos(x), x)
```

```python
sp.integrate(sp.exp(-x), (x, 0, sp.oo)) # 0부터 무한대까지 적분
```

```python
sp.integrate(sp.exp(-x**2 - y**2), (x, -sp.oo, sp.oo), (y, -sp.oo, sp.oo))
```

### 다변수함수의 극점, 안장점 구하기
- `sympy.solve()` 함수를 통해 방정식의 해를 구할 수 있음

```python
expr = 3 * x - x ** 3 - 2 * y ** 2 + y ** 4
expr
```

```python
sp.solve(sp.diff(expr, x))
```

```python
sp.solve(sp.diff(expr, y))
```

#### Hessian 구하기
- `sp.hessian()` 사용
- ref: https://docs.sympy.org/latest/modules/matrices/matrices.html#sympy.matrices.dense.hessian

```python
sp.hessian(expr, (x, y))
```

- det() 를 통해 행렬의 행렬식을 구할 수 있음

```python
hess = sp.hessian(expr, (x, y)).det()
hess
```

```python
hess.subs([(x, -1), (y, -1)])
```

```python
d2 = sp.diff(expr, x, 2)
d2.subs([(x, -1), (y, -1)])
```

```python
hess.subs([(x, -1), (y, 0)])
```
